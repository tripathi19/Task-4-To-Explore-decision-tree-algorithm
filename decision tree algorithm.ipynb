# importing packages and the Iris dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
iris=pd.read_csv("C:/Users/ABHISHEK TRIPATHI/Downloads/Iris.csv")
iris
```

# preparing dataset

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
iris['Species']=le.fit_transform(iris['Species'])
```

```python
z=iris.iloc[:,[1,2,3]]
v=iris.iloc[:,-1]
```

```python
#splitting the data
from sklearn.model_selection import train_test_split
(z_train,z_test,v_train,v_test)=train_test_split(z,v,test_size=0.3,random_state=101)
```

# fitting the model

```python
from sklearn.tree import DecisionTreeClassifier
classifier=DecisionTreeClassifier(random_state=0)
classifier.fit(z_train,v_train)
```

# Prediction and Evaluation

```python
v_pred=classifier.predict(z_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
confusion_matrix(v_test,v_pred)
```

```python
accuracy_score(v_test,v_pred)
```

# Decision Tree Algorithm

```python
from sklearn import tree
plt.figure(figsize=(16,10))
tree.plot_tree(classifier,filled=True)
plt.title("Decision Tree Diagram")
```
